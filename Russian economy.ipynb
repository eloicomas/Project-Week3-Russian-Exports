#Evolution of trading with the top 20 partners.

```python
import pandas as pd

#The value/year that Russia exports for each country

amount_exports_country = pd.pivot_table(
    data=df_exports,
    values='Trade Value (US$)',
    index='Year',
    columns='Partner',
    aggfunc='sum'
)
amount_exports_country


```

```python
sum_exports=amount_exports_country.sum()
sum_exports
```

```python
#Sum of exports during the period 2007-2020, the 20 highest values from russian partners.
top_25_total=sum_exports.nlargest(25)
top_25_total
```

### Visualization



```python
import matplotlib.pyplot as plt

x = top_25_total.index
y = top_25_total.values


plt.bar(x, y, color='blue')
plt.xlabel('Countries')
plt.ylabel('Total exports')
plt.title('Top 20 countries with highest exports')


plt.xticks(rotation=45, ha='right')


plt.tight_layout()
plt.show()

```

```python
mean_exports=amount_exports_country.mean()
mean_exports
```

```python
#Mean of exports during the period 2007-2020, the 20 highest values from russian partners.
top_25_mean=mean_exports.nlargest(25)
top_25_mean
```

```python
x = top_25_mean.index
y = top_25_mean.values


plt.bar(x, y, color='red')
plt.xlabel('Countries')
plt.ylabel('Total exports')
plt.title('Mean exports with countries of higher total amount ')


plt.xticks(rotation=45, ha='right')


plt.tight_layout()
plt.show()
```

```python
#Check if the stats are consistent
a=top_25_total.index.tolist()
print(a)

b=top_25_mean.index.tolist()
print(b)



```

```python
#Evolution of the exports 

import pandas as pd
import matplotlib.pyplot as plt



partner_totals = amount_exports_country.sum()

top_15_partners = partner_totals.nlargest(15).index

filtered_pivot = amount_exports_country[top_15_partners]


plt.figure(figsize=(10, 6))
for partner in filtered_pivot.columns:
    plt.plot(filtered_pivot.index, filtered_pivot[partner], label=partner)

plt.xlabel("Year")
plt.ylabel("Trade Value (US$)")
plt.title("Trade Value by Partner (Top 15)")
plt.legend()
plt.grid(True)
plt.show()
```
